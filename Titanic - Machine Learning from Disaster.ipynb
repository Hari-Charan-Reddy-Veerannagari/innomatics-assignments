Cabin has nulls of more than 60%. So, lets drop the entire column

```
data.drop('Cabin', axis= 1, inplace= True)
data
```

```
numeric_data = data.select_dtypes(include = 'number')

for column in numeric_data.columns:
    data[column].fillna(value = data[column].median(),inplace = True)
```

```
object_data = data.select_dtypes(exclude = 'number')

for column in object_data.columns:
    data[column].fillna(value = data[column].mode(),inplace = True)
```
