```
import requests
import re
from bs4 import BeautifulSoup
import js2xml
```

```
response = requests.get('https://www.worldometers.info/coronavirus/country/india/')
```

```
soup = BeautifulSoup(response.content)
```

```
for i in soup.find_all('script',  string=re.compile("Highcharts.chart")):
    parsed = js2xml.parse(i.text)
    print(js2xml.pretty_print(parsed))
```
